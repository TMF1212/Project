```python
import pandas as pd

# Identify Agentic PRs
# Assuming 'agent' column exists and contains agent identifiers or usernames
agent_candidates = ['OpenAI Codex', 'Devin', 'GitHub Copilot', 'Cursor', 'Claude Code']
agent_prs = all_pr_df[all_pr_df['agent'].isin(agent_candidates)]
```

```python
#  Link PRs with user data to get user creation dates
# Merge PRs with users on user_id
agent_prs_users = agent_prs.merge(all_user_df, left_on='user_id', right_on='id', how='left')
```

```python
# Convert relevant date columns to datetime
agent_prs_users['pr_created_at'] = pd.to_datetime(agent_prs_users['created_at_x'])  # PR creation date
agent_prs_users['user_created_at'] = pd.to_datetime(agent_prs_users['created_at_y'])  # User account creation
```

```python
# Calculate user experience at the time of PR creation
agent_prs_users['experience_days'] = (agent_prs_users['pr_created_at'] - agent_prs_users['user_created_at']).dt.days
```

```python
# Classify developers into newcomers and experienced
# Newcomers: less than 180 days (6 months)
newcomers = agent_prs_users[agent_prs_users['experience_days'] < 180]
```

```python
# Experienced: at least 365 days (1 year)
experienced = agent_prs_users[agent_prs_users['experience_days'] >= 365]
```

```python
# Step 6: Analyze the adoption pattern
num_newcomers = len(newcomers)
num_experienced = len(experienced)
# Calculate the number of unique users who used agents
num_unique_agent_users = agent_prs_users['user_id'].nunique()
```

```python
#  Cross-tabulate by repository or other metadata for ecosystem analysis
# Merge with all_repo_df to get repository names
agent_prs_users_with_repo = agent_prs_users.merge(all_repo_df, left_on='repo_id', right_on='id', how='left')

# group by repository full name
adoption_by_repo = agent_prs_users_with_repo.groupby('full_name')['user_id'].nunique()

# Output results
print(f"Total Agentic PRs analyzed: {len(agent_prs)}")
print(f"Number from newcomers (<6 months): {num_newcomers}")
print(f"Number from experienced developers (>=1 year): {num_experienced}")
print(f"Number of unique users who used agents: {num_unique_agent_users}")
print("\nAdoption by repository (number of unique users who used agents):")
print(adoption_by_repo)
```

```python
total_prs = len(agent_prs)
proportion_newcomers = num_newcomers / total_prs
proportion_experienced = num_experienced / total_prs

print(f"Proportion of agent PRs from newcomers: {proportion_newcomers:.2%}")
print(f"Proportion from experienced developers: {proportion_experienced:.2%}")


```

```python
import matplotlib.pyplot as plt

# Data for pie chart
labels = ["Newcomers (<6 months)", "Experienced (>=1 year)"]
sizes = [num_newcomers, num_experienced]

# Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=["#66b3ff", "#99ff99"]
)
plt.title("Adoption of Coding Agents: Newcomers vs Experienced")
plt.show()
```

```python
# Count Agentic PRs per repository
adoption_by_repo = all_pr_df.groupby("repo_id")["agent"].apply(lambda x: (x == "Claude_Code").sum())
adoption_by_repo = adoption_by_repo.sort_values(ascending=False)

print("Top repositories by Agentic PR adoption:")
print(adoption_by_repo.head(10))
```

```python
# Total PRs per repo
total_prs_by_repo = all_pr_df.groupby("repo_id")["id"].count()

# Fraction of PRs that are agentic
adoption_fraction = adoption_by_repo / total_prs_by_repo
adoption_fraction = adoption_fraction.sort_values(ascending=False)

print("Repos with highest proportion of Agentic PRs:")
print(adoption_fraction.head(10))
```

```python
import matplotlib.pyplot as plt

# Plot top 10 repos by proportion
top10_fraction = adoption_fraction.head(10)

plt.figure(figsize=(10, 6))
top10_fraction.plot(kind="bar", color="skyblue")
plt.ylabel("Proportion of Agentic PRs")
plt.title("Top 10 Repositories by Agent Adoption")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Extract ecosystem (repo_url format: https://api.github.com/repos/{owner}/{repo})
all_pr_df["ecosystem"] = all_pr_df["repo_url"].apply(lambda x: x.split("/")[-2])

# Count agentic PRs per ecosystem
adoption_by_ecosystem = all_pr_df.groupby("ecosystem")["agent"].apply(lambda x: (x == "Claude_Code").sum())
total_by_ecosystem = all_pr_df.groupby("ecosystem")["id"].count()

# Fraction
ecosystem_fraction = (adoption_by_ecosystem / total_by_ecosystem).sort_values(ascending=False)

print("Ecosystem-level adoption patterns:")
print(ecosystem_fraction.head(10))
```

```python
# Top ecosystems by proportion
top10_ecosystem = ecosystem_fraction.head(10)

plt.figure(figsize=(10, 6))
top10_ecosystem.plot(kind="bar", color="lightgreen")
plt.ylabel("Proportion of Agentic PRs")
plt.title("Top 10 Ecosystems by Agent Adoption")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
# Mark whether PR is agentic (submitted by an AI agent like Claude_Code, Devin, etc.)
all_pr_df["is_agentic"] = all_pr_df["agent"].notnull()
```

```python
# PR size = number of characters in body
all_pr_df["pr_size"] = all_pr_df["body"].fillna("").apply(len)

# Compare average PR size
avg_size_agentic = all_pr_df[all_pr_df["is_agentic"]]["pr_size"].mean()
avg_size_human   = all_pr_df[~all_pr_df["is_agentic"]]["pr_size"].mean()

print("Average PR size (chars) - Agentic:", avg_size_agentic)
print("Average PR size (chars) - Human:", avg_size_human)
```

```python
def classify_task(title):
    title = str(title).lower()
    if "fix" in title or "bug" in title:
        return "bugfix"
    elif "feat" in title or "add" in title or "implement" in title:
        return "feature"
    elif "refactor" in title:
        return "refactor"
    elif "doc" in title or "readme" in title:
        return "docs"
    else:
        return "other"

# Apply task classification
all_pr_df["task_type"] = all_pr_df["title"].apply(classify_task)

# Count distribution
task_counts = all_pr_df.groupby(["task_type", "is_agentic"])["id"].count().unstack(fill_value=0)
print(task_counts)
```

```python
if "commits" in all_pr_df.columns:
    avg_commits_agentic = all_pr_df[all_pr_df["is_agentic"]]["commits"].mean()
    avg_commits_human   = all_pr_df[~all_pr_df["is_agentic"]]["commits"].mean()

    print("Average commits per PR - Agentic:", avg_commits_agentic)
    print("Average commits per PR - Human:", avg_commits_human)
else:
    print("No 'commits' column found in dataset â€” skipping commit granularity.")
```
